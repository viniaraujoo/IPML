```python
# Auxiliar libs
import sys
import os
import urllib.request
import tarfile
import zipfile
import collections

# Manipulate data
import numpy as np
import pandas as pd


import h5py 
import keras
from keras.applications import inception_v3 as inc_net

# Display
from IPython.display import Image
import matplotlib.pyplot as plt

np.random.seed(1)  # makes the results reproducible
```

```python
inet_model = inc_net.InceptionV3()
```

```python
import requests

# you can change this URL to any image URL you want!
IMAGE_URL = 'https://raw.githubusercontent.com/marcotcr/lime/master/doc/notebooks/data/cat_mouse.jpg' 
img_data = requests.get(IMAGE_URL).content
```

```python
'''
import module impl
parameters: 
            + model_predict: classifier prediction probability function, which
                             takes a numpy array and outputs prediction probabilities.  For
                             ScikitClassifiers , this is classifier.predict_proba.
            + test: image test model
'''
import impl.impl as ts
exemple = ts.impl(model_predict=inet_model.predict,test=img_data)
```

```python
# Example of function return considering the main points of the image
# by default top labels equal to 5 then n goes from 0 to 4 for the exhition
plt.imshow(exemple[2])
```
